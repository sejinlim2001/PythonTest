```python
import csv
import matplotlib.pyplot as plt

f = open('deajeon_temp.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data)
high = []
low = []

for row in data:
    if row[-1] != '' and row[-2] != '':
        date = row[0].split('-')
        if date[1] == '12' and date[2] == '27':
            high.append(float(row[-1]))
            low.append(float(row[-2]))

plt.rcParams['axes.unicode_minus'] = False
plt.title('Birthday temperature change graph')
plt.plot(high, 'hotpink', label='high')
plt.plot(low, 'skyblue', label='low')
plt.legend()
plt.show()
```
