```python
import csv
import matplotlib.pyplot as plt

f = open('Daejeon_temp.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)

print(header)

avg_temp = []
for i in range(31):
    avg_temp.append([])

for row in data:
    if row[2] != '':
        date = row[0].split('-')
        if date[1] == '08':
            avg_temp[int(date[2])-1].append(float(row[2]))

f.close()

plt.style.use('ggplot')
plt.figure(figsize=(10, 5), dpi=300)
plt.title('Daejeon daily average temperature distribution')
plt.boxplot(avg_temp, showfliers=False)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('Daejeon_temp.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)

print(header)

january = []
august = []

for row in data:
    if row[-1] != '':
        date = row[0].split('-')
        if date[1] == '01':
            january.append(float(row[-1]))
        if date[1] == '08':
            august.append(float(row[-1]))

f.close()

plt.title('Daejeon January and August Highest Temperature Frequency')
plt.hist(january, color='b', bins=100, label='jan')
plt.hist(august, color='r', bins=100, label='aug')
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('Daejeon_temp.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)

print(header)

high_temp = []
low_temp = []

for row in data:
    if row[-1] != '' and row[-2] != '':
        date = row[0].split('-')
        if date[1] == '12' and date[2] == '27':
            high_temp.append(float(row[-1]))
            low_temp.append(float(row[-2]))

f.close()

plt.title('Temperature change graph on my birthday')
plt.plot(high_temp, 'hotpink', label='high')
plt.plot(low_temp, 'skyblue', label='low')
plt.legend()
plt.show()
```
